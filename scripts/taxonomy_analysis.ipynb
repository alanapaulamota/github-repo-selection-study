```python
import pandas as pd
import matplotlib.pyplot as plt
import os


# ==========================
# Configurações
# ==========================

INPUT_FILE = "taxonomy_spm.csv"
OUTPUT_DIR = "taxonomy_analysis"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ==========================
# Carregar dados
# ==========================

df = pd.read_csv(INPUT_FILE)

print("Total de registros:", len(df))


# ==========================
# Estatísticas gerais
# ==========================

total_codes = len(df)
total_groups = df["Group"].nunique()
total_themes = df["Themes"].nunique()

print("\nResumo:")
print("Códigos:", total_codes)
print("Grupos:", total_groups)
print("Temas:", total_themes)


# ==========================
# Distribuição por grupo
# ==========================

group_count = df["Group"].value_counts()

group_count.to_csv(
    f"{OUTPUT_DIR}/group_distribution.csv"
)


plt.figure(figsize=(10,6))

group_count.sort_values().plot(
    kind="barh"
)

plt.title(
    "Distribuição de códigos por Grupo"
)

plt.xlabel(
    "Quantidade"
)

plt.tight_layout()

plt.savefig(
    f"{OUTPUT_DIR}/01_groups_distribution.png",
    dpi=300
)

plt.close()



# ==========================
# Distribuição por tema
# ==========================

theme_count = df["Themes"].value_counts()

theme_count.to_csv(
    f"{OUTPUT_DIR}/theme_distribution.csv"
)


plt.figure(figsize=(10,6))

theme_count.sort_values().plot(
    kind="barh"
)

plt.title(
    "Distribuição por Tema"
)

plt.xlabel(
    "Quantidade"
)

plt.tight_layout()

plt.savefig(
    f"{OUTPUT_DIR}/02_themes_distribution.png",
    dpi=300
)

plt.close()



# ==========================
# Top códigos
# ==========================

code_count = (
    df["Codes"]
    .value_counts()
)


code_count.head(20).to_csv(
    f"{OUTPUT_DIR}/top_codes.csv"
)


plt.figure(figsize=(10,8))


code_count.head(20).sort_values().plot(
    kind="barh"
)


plt.title(
    "Top 20 códigos mais frequentes"
)

plt.xlabel(
    "Quantidade"
)

plt.tight_layout()


plt.savefig(
    f"{OUTPUT_DIR}/03_top_codes.png",
    dpi=300
)

plt.close()



# ==========================
# Relação Grupo x Tema
# ==========================

cross = pd.crosstab(
    df["Group"],
    df["Themes"]
)


cross.to_csv(
    f"{OUTPUT_DIR}/group_theme_matrix.csv"
)



plt.figure(figsize=(12,8))


plt.imshow(
    cross,
    aspect="auto"
)


plt.xticks(
    range(len(cross.columns)),
    cross.columns,
    rotation=90
)


plt.yticks(
    range(len(cross.index)),
    cross.index
)


plt.title(
    "Relação Grupo × Tema"
)


plt.colorbar()


plt.tight_layout()


plt.savefig(
    f"{OUTPUT_DIR}/04_group_theme_heatmap.png",
    dpi=300
)


plt.close()



# ==========================
# Relatório Markdown
# ==========================


with open(
    f"{OUTPUT_DIR}/taxonomy_report.md",
    "w",
    encoding="utf-8"
) as file:


    file.write(
        "# Relatório da Taxonomia de Structural Pattern Matching\n\n"
    )


    file.write(
        "## Visão Geral\n\n"
    )


    file.write(
        f"- Total de códigos: {total_codes}\n"
    )

    file.write(
        f"- Total de grupos: {total_groups}\n"
    )

    file.write(
        f"- Total de temas: {total_themes}\n\n"
    )



    file.write(
        "## Grupos mais frequentes\n\n"
    )


    for name,value in group_count.items():

        percentage = (
            value / total_codes
        ) * 100

        file.write(
            f"- {name}: {value} ({percentage:.2f}%)\n"
        )



    file.write(
        "\n## Temas mais frequentes\n\n"
    )


    for name,value in theme_count.items():

        percentage = (
            value / total_codes
        ) * 100

        file.write(
            f"- {name}: {value} ({percentage:.2f}%)\n"
        )



    file.write(
        "\n## Principais códigos\n\n"
    )


    for name,value in code_count.head(20).items():

        file.write(
            f"- {name}: {value}\n"
        )



    file.write(
        "\n## Insights iniciais\n\n"
    )


    file.write(
        f"- O grupo dominante foi **{group_count.idxmax()}**.\n"
    )


    file.write(
        f"- O tema dominante foi **{theme_count.idxmax()}**.\n"
    )


    file.write(
        "- A concentração dos códigos indica quais aspectos do Structural Pattern Matching receberam maior atenção durante a definição da linguagem.\n"
    )


print("\nAnálise finalizada!")
print(f"Arquivos salvos em: {OUTPUT_DIR}")
```

```python
from PIL import Image
import os

# Pasta onde estão as imagens geradas
INPUT_DIR = "taxonomy_analysis"

# Lista todos os PNGs da pasta
files = [
    f for f in os.listdir(INPUT_DIR)
    if f.endswith(".png")
]

for file in files:
    png_path = os.path.join(INPUT_DIR, file)

    # Nome do PDF
    pdf_path = os.path.join(
        INPUT_DIR,
        file.replace(".png", ".pdf")
    )

    # Abre imagem
    img = Image.open(png_path)

    # Garante modo RGB (necessário para PDF)
    img = img.convert("RGB")

    # Salva como PDF
    img.save(pdf_path)

    print(f"Convertido: {pdf_path}")

print("Todos os gráficos foram convertidos para PDF!")
```
